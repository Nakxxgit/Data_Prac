```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
a = pd.ExcelFile("전국무료와이파이표준데이터.xls")
df = a.parse()
```

```python
Seoul_wifi = df[df["설치시도명"] == "서울특별시"]
Seoul_wifi
```

```python
map_Seoul = folium.Map(location=[37.6,127], zoom_start=10)
my_marker_Cluster = MarkerCluster()
```

```python
name_lst = Seoul_wifi["설치장소명"].tolist()
op_lst = Seoul_wifi["서비스제공사명"].tolist()
SSID_lst = Seoul_wifi["와이파이SSID"].tolist()
lat_lst = Seoul_wifi["위도"].tolist()
lng_lst = Seoul_wifi["경도"].tolist()
```

```python
m = folium.Map([43, -100], zoom_start=4)
html = "<h3>" + SSID_lst[0] + "</h3><br>" + op_lst[0]
```

```python
for name, op, SSID, lat, lng in zip(name_lst, op_lst, SSID_lst, lat_lst, lng_lst):
    folium.Marker([lat_lst[0], lng_lst[0]], popup=html).add_to(my_marker_Cluster)
```

```python
map_Seoul.add_child(my_marker_Cluster)
```

```python
map_Seoul.save(outfile = "test.html")
```
